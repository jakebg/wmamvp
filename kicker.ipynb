```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm, colors
import numpy.random
import seaborn as sns
```

```python
kicker = pd.read_csv('stats/nfl/KICKER.csv')
print(kicker.Player.value_counts())
print('Kicker', kicker.Player.value_counts().size)
```

```python
fig, ax = plt.subplots()
kicker.plot('FGM', 'XPM', kind='scatter', ax=ax, color='#e9dbf9')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

x = 0
for k,v  in kicker.iterrows():
    name = v['Player']
    oi = v['FGM']
    yoi = v['XPM']
    v = (oi,yoi)
    ax.annotate(name,v,xytext=(10,-5), textcoords='offset points',
                family='sans-serif', fontsize=8)
    x+=1
    if x > 5:
        break


ax.set_xlabel('FGM')
ax.set_ylabel('XPM')
```

```python
fig, ax = plt.subplots()
kicker.plot('Year', 'Pts', kind='scatter', ax=ax, color='#e9dbf9')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

x = 0
for k,v  in kicker.iterrows():
    name = v['Player']
    oi = v['Year']
    yoi = v['Pts']
    v = (oi,yoi)
    ax.annotate(name,v,xytext=(10,-5), textcoords='offset points',
                family='sans-serif', fontsize=8)
    x+=1
    if x > 5:
        break


ax.set_xlabel('Year')
ax.set_ylabel('Pts')
```

```python
kicker_year = kicker[(kicker['Year'] == 2011) | (kicker['Year'] == 1998)|(kicker['Year'] == 2003)|(kicker['Year'] == 1983) | (kicker['Year'] == 2016)].copy()

kicker_year
```

```python
kicker_year.boxplot(by='Year', column=['Pts'],patch_artist=True,whis=[0,95],  flierprops=dict(color='red', markeredgecolor='purple'))

plt.title('')
ax.set_xlabel('')
ax.set_ylabel('TPts')
plt.show()
```
